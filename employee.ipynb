# Employee Dataset Visualization & Analysis

```python
#How many employees are there in each department? 
import pandas as pd
df=pd.read_csv(r"C:/Users/parth/OneDrive/Desktop/pandas/raw_employee_data_one_row.csv",header=None)
columns=['Name','Employee_Id','Age','Gender','Department','Salary','Address']

data_list=df.values.flatten().tolist()
records=[data_list[i:i+7] for i in range(0,len(data_list),7)]
df_clean=pd.DataFrame(records,columns=columns)

total_employee=df_clean.groupby('Department')['Employee_Id'].count()
print(total_employee)
```

```python
#What is the average salary of employees in each department? 
avg_salary=df_clean.groupby('Department')['Salary'].mean()
print(avg_salary)
```

```python
#List all employees aged between 30 and 40 working in Engineering. 
lst=df_clean[(df_clean['Age'] >= 30) & (df_clean['Age'] <= 40) & (df_clean['Department'] == 'Engineering')]
print('list of all employees aged between 30and 40 working in engineering is:\n',lst)
```

```python
#What is the standard deviation of employee salaries?
std_dv=df_clean['Salary'].std()
print('standard daviation of employee salaries will be:',std_dv)
```

```python
#How many employees earn a salary above the overall mean salary?

mean_salary=df_clean['Salary'].mean()
a=df_clean[df_clean['Salary']>mean_salary].shape[0]
print('number of employees earn salary above mean salary is:',a)
```

```python
#Create a NumPy array of unique employee ages and sort it.
import numpy as np
age=df_clean['Age']
sorted_age= np.sort(age.unique())
print(sorted_age)
```

```python
# Create a bar chart of the number of employees in each department. 
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("C:/Users/parth/OneDrive/Desktop/pandas/raw_employee_data_one_row.csv",header=None)
columns=['Name','Employee_Id','Age','Gender','Department','Salary','Address']

data_list=df.values.flatten().tolist()
records=[data_list[i:i+7] for i in range(0,len(data_list),7)]
df_clean=pd.DataFrame(records,columns=columns)

total_employee=df_clean.groupby('Department')['Employee_Id'].count()
colors=['red','lightblue','lavender','yellow','lightgreen','lightpink']
plt.bar(total_employee.index,total_employee.values,color=colors)
plt.title('Number of employee in each department')
plt.xlabel('Department')
plt.ylabel('Number of employee')
plt.show()
```

```python
#Plot a histogram of employee ages. 
a=df_clean['Age']
plt.hist(a,bins=15,color='gray',edgecolor='white')
plt.xlabel('Age')
plt.title('Distribution of Employee Ages')
plt.show()
```

```python
#Create a pie chart of gender distribution.
import matplotlib.pyplot as plt
category=df_clean['Gender'].value_counts()
ex=[0,0,0.01]
plt.pie(category,labels=category.index,autopct='%1.1f%%',explode=ex)
plt.title('Gender Distribution')
plt.show()
```

```python
#Plot a scatter plot between Age and Salary to identify any patterns

import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_csv("C:/Users/parth/OneDrive/Desktop/pandas/raw_employee_data_one_row.csv",header=None)
columns=['Name','Employee_Id','Age','Gender','Department','Salary','Address']

data_list=df.values.flatten().tolist()
records=[data_list[i:i+7] for i in range(0,len(data_list),7)]
df_clean=pd.DataFrame(records,columns=columns)

plt.scatter(df_clean['Age'],df_clean['Salary'],marker='*',color='black')
plt.show()
```
